Like it was said before, the dataset contains four files divided by victim count. 

I want to join them into one file using glob and os (it will find all files that matches the concrete pathname - in this case "victim_count") and pd.concat function.

```python
data_files = glob.glob(os.path.join('', './*victim_count.csv'))
df = pd.concat(map(lambda file: pd.read_csv(file), data_files),ignore_index=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

I've got 305 rows and 6 columns in our dataframe and I have missing values (68) in only one column ("Possible victims") - so far, so good. 

First and foremost I would like to clear out the "Proven victims" and "Possible victims" column.

It came to my attention almost immediately that these two columns can provide me with some significant insights and interesting analysis results. 

And from the first look at the "Possible victims" data I can tell that there is a lot of work to do.

```python
df['Possible victims'].value_counts(dropna=False)
```

Above function let me group the values and show them in a more see-through way.

I can see that I have a lot of missing values ('NaN') in the dataset in this particular column - 68 to be exact. 

I have also values with different characters added to integers like '+', '?' or "~". 

I can see few spots where the value is described as "Unknown" or "-".

There is also quite a lot of values in intervals. 

The "Possible victims" column even from a name is indicating that the data in it is speculative. 

I will make a few assumptions and clean the data in accordance to them:

- the 'Nan' values will be replaced with corresponding values from the "Proven victims" column,
- the "Unknown" and "-" values will be replaced with corresponding values from the "Proven victims" column, 
- when it comes to values in intervals, I will leave the highest number from the interval.
- the values with '+', '?' or "~", I will erase the special characters and assume the integer as a correct value,


Let's localize missing values in "Possible victims" and display them alongside the "Proven victims" column - I already know that the missing values only occur in the "Possible victims" column.

```python
df.loc[df['Possible victims'].isna(), ['Proven victims','Possible victims']] 
```

Now, I will fill up the missing values with the values from corresponding rows in "Proven victims" column.

```python
df['Possible victims'] = df['Possible victims'].fillna(df['Proven victims'])
```

```python
df.isnull().sum()
```

It worked - there is no missing values in the "Possible victims" column. 

Now, I will go right into my next task - replacing the "Unknown" and "-" values with corresponding values from the "Proven victims" column.

Let's localize these values in "Possible victims" and display them alongside the "Proven victims" column.

```python
df.loc[df['Possible victims'].isin(['Unknown', '-']), ['Name','Proven victims','Possible victims']] 
```

Next step is filling up the values with "Unknown" and "-" from "Possible victims" with the values from corresponding rows from the "Proven victims" column.

```python
df.loc[(df['Possible victims'].isin(['Unknown', '-']))&(df['Proven victims'].notnull()), 'Possible victims'] = df['Proven victims']
```

Let's check if the filling up was carried out correctly for one individual - for example Yoshio Kodaira. 

```python
df[df['Name'].str.contains('Yoshio Kodaira')]
```

I can see that the value from the "Proven victims" column was copied into "Possible victims" column for that row. 

Now, I will clean up the values in intervals. 

From what I saw intervals consist of integer, dash sign and other integer, let's display these lines.

```python
df[df['Possible victims'].str.contains('-') == True]
```

There was definitely a lot more values in intervals when I carried out 'values.count' function. 

I went back to check it and it seems that there is a longer dash used in few of the intervals.

Let's display these lines.

```python
df[df['Possible victims'].str.contains('–') == True]
```

It seems that now I have all the lines I need. 

Now, let's do the split for each sign so I can have only the highest number from interval in the "Possible victims" column. 

```python
df['Possible victims'] = [str(x).split('-')[-1] for x in df['Possible victims']]
df['Possible victims'] = [str(x).split('–')[-1] for x in df['Possible victims']]
```

I will check if I am left with any values in intervals for "Possible victims" column.

```python
df['Possible victims'].value_counts(dropna=False)
```

I am not! That's great news. 

The only thing that I am left to clean in this particular column are the special characters '+', '?' or "~". 

```python
df["Possible victims"] = df["Possible victims"].str.replace("+","")
df["Possible victims"] = df["Possible victims"].str.replace("?","")
df["Possible victims"] = df["Possible victims"].str.replace("~","")
```

Let's check if all the values in the "Possible victims" column are cleaned up.

```python
df['Possible victims'].value_counts(dropna=False)
```

It's all cleaned up! 

I have managed to do that without erasing any row of the data. 

I assumed it would be a good choice due to a speculative nature of the data in this particular column. 

Now, let's check the "Proven victims" column.

```python
df['Proven victims'].value_counts(dropna=False)
```

Right of the bat, I can see that the data is way cleaner that in the "Possible victims" column.

However, I still have some values in intervals and one "+" character beside integer despite the fact that the data in this column should be a confirmed and concrete one. 

It may be due to many things - inconsistencies in the sources, lack of or incomplete documentation of the trials for each murder etc. 

I can either erase the rows with such values or include them.

I decided to include them with below assumptions: 

- when it comes to values in intervals, I will leave the highest number from the interval,
- the values with '+' - I will erase the special character and assume the integer as a correct value.

Let's check the values in intervals.

I will check both signs - small dash and longer dash - like in "Possible victims" column.

```python
df[df['Proven victims'].str.contains('-') == True]
```

```python
df[df['Proven victims'].str.contains('–') == True]
```

It seems that there are only intervals with longer dash. 

Let's split the values up and leave the highest one from interval.

```python
df['Proven victims'] = [str(x).split('–')[-1] for x in df['Proven victims']]
```

I will erase the "+" sign now. 

```python
df["Proven victims"] = df["Proven victims"].str.replace("+","")
```

I will check if there is anything else to clean in this column.

```python
df['Proven victims'].value_counts(dropna=False)
```

It seems not! The "Proven victims" and "Possible victims" column are all good and clean.
